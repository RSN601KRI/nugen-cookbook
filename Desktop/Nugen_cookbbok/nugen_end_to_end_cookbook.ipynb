## Preparing the Nugen API Client

At the time of writing, Nugen APIs are accessed via HTTP endpoints.
Below, we define a minimal client wrapper to keep the notebook clean and reusable.

```python
import requests
import json

class NugenClient:
    def __init__(self, api_key: str):
        self.api_key = api_key
        self.base_url = "https://api.nugen.in/api/v3/inference"

    def generate_text(self, model, prompt, max_tokens=200, temperature=0.7):
        payload = {
            "model": model,
            "prompt": prompt,
            "max_tokens": max_tokens,
            "temperature": temperature
        }

        response = requests.post(
            f"{self.base_url}/generate",
            headers={
                "Authorization": f"Bearer {self.api_key}",
                "Content-Type": "application/json"
            },
            data=json.dumps(payload)
        )

        response.raise_for_status()
        return response.json()
```

## Step 6: Handling Errors Gracefully

Production-grade AI systems must handle failures explicitly.
Instead of failing silently, applications should surface errors clearly
and allow developers to debug issues efficiently.

```python
def safe_generate(model_name, prompt):
    if model_name != "nugen-text-1":
        return {"error": "Invalid or unsupported model"}
    if not prompt or prompt.strip() == "":
        return {"error": "Prompt cannot be empty"}
    return simulated_response

error_example = safe_generate("invalid-model", "Test")
print(error_example)
```

This approach ensures:

- Failures are visible
- Applications donâ€™t crash silently
- Debugging remains straightforward
